<h1>Remove Comments</h1><p>Given a C++ program, remove comments from it. The program <code>source</code> is an array where <code>source[i]</code> is the <code>i</code>-th line of the source code.  This represents the result of splitting the original source code string by the newline character <code>\n</code>.</p><div></div><div></div><p>In C++, there are two types of comments, line comments, and block comments.</p><div></div><p><div></div>The string <code>//</code> denotes a line comment, which represents that it and rest of the characters to the right of it in the same line should be ignored.<div></div></p><p><div></div>The string <code>/*</code> denotes a block comment, which represents that all characters until the next (non-overlapping) occurrence of <code>*/</code> should be ignored.  (Here, occurrences happen in reading order: line by line from left to right.)  To be clear, the string <code>/*/</code> does not yet end the block comment, as the ending would be overlapping the beginning.<div></div></p><p><div></div>The first effective comment takes precedence over others: if the string <code>//</code> occurs in a block comment, it is ignored. Similarly, if the string <code>/*</code> occurs in a line or block comment, it is also ignored.<div></div></p><p><div></div>If a certain line of code is empty after removing comments, you must not output that line: each string in the answer list will be non-empty.<div></div></p><p><div></div>There will be no control characters, single quote, or double quote characters.  For example, <code>source = "string s = "/* Not a comment. */";"</code> will not be a test case.  (Also, nothing else such as defines or macros will interfere with the comments.)<div></div></p><p><div></div>It is guaranteed that every open block comment will eventually be closed, so <code>/*</code> outside of a line or block comment always starts a new comment.<div></div></p><p><div></div>Finally, implicit newline characters can be deleted by block comments. The string <code>//</code> denotes line 4 as comments.<div></div></pre><div></div></p><div></div><div></div><p><b>Example 2:</b><br /><div></div><pre style="white-space: pre-wrap"><div></div><b>Input:</b> <div></div>source = ["a/*comment", "line", "more_comment*/b"]<div></div><b>Output:</b> ["ab"]<div></div><b>Explanation:</b> The original source string is "a/*comment<b>\n</b>line<b>\n</b>more_comment*/b", where we have bolded the newline characters.
